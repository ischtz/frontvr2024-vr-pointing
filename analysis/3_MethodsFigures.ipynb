# Methods Figures

```python
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
FOLDER_IMG = '../figures/images'
FOLDER_FIG = '../figures'

# Object 3x2 figure
IMGS = np.array([["oben_sym.png", "oben_asym.png"],
                 ["mitte_sym.png", "mitte_asym.png"],
                 ["unten_sym.png", "unten_asym.png"]])

# Task figure
TASKS = np.array(["baseline_sphere.png", "object_hand.png", "pointing_calib.png", "pointing_tracker.png"])
TASK_LABELS = ['A. Baseline Task (Sphere)', 'B. Object Task (Hand)', 'C. Vive Tracker Calibration', 'D. Pointing Example']

# Figure styling
LABEL_SIZE = 14
LABEL_PAD = 10
OUTLINE_WIDTH = 1.5
```

```python
# Object Figure - 2x3 column version
fig, ax = plt.subplots(IMGS.shape[0], IMGS.shape[1], figsize=(5, 10), layout='compressed')

obj_num = 1
for row in range(0, IMGS.shape[0]):
    for col in range(0, IMGS.shape[1]):

        img = plt.imread(os.path.join(FOLDER_IMG, IMGS[row][col]))
        ax[row, col].imshow(img)
        ax[row, col].grid(False)
        
        #ax[row, col].axvline(img.shape[0]/2, color='k', ls='--', lw=0.6)
        
        #ax[row, col].text(0.03, 0.9, str(obj_num), transform = ax[row, col].transAxes, horizontalalignment='left', color='k', fontsize=LABEL_SIZE)

        obj_num += 1

# Applies to all subplots
for axs in ax.flatten():
        axs.set_xticklabels([])
        axs.set_xticks([])
        axs.set_yticks([])
        axs.set_yticklabels([])

ax[0][0].set_ylabel('High', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)
ax[1][0].set_ylabel('Middle', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)
ax[2][0].set_ylabel('Low', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)
ax[0][0].xaxis.set_label_position('top') 
ax[0][1].xaxis.set_label_position('top') 
ax[0][0].set_xlabel('Symmetric', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)
ax[0][1].set_xlabel('Asymmetric', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)

#plt.savefig(os.path.join(FOLDER_FIG, 'objects_col.png'), dpi=300, bbox_inches='tight')
```

```python
# Object Figure - Space-saving 3x2 row version
fig, ax = plt.subplots(IMGS.shape[1], IMGS.shape[0], figsize=(10, 5), layout='compressed')

obj_num = 1
for col in range(0, IMGS.shape[0]):
    for row in range(0, IMGS.shape[1]):

        img = plt.imread(os.path.join(FOLDER_IMG, IMGS[col][row]))
        ax[row, col].imshow(img)
        ax[row, col].grid(False)
        
        #ax[row, col].axvline(img.shape[0]/2, color='k', ls='--', lw=0.6)
        
        #ax[row, col].text(0.03, 0.9, str(obj_num), transform = ax[row, col].transAxes, horizontalalignment='left', color='k', fontsize=LABEL_SIZE)

        obj_num += 1

# Applies to all subplots
for axs in ax.flatten():
    axs.set_xticklabels([])
    axs.set_xticks([])
    axs.set_yticks([])
    axs.set_yticklabels([])
        
    for axis in ['top','bottom','left','right']:
        axs.spines[axis].set_linewidth(OUTLINE_WIDTH)

ax[0][0].xaxis.set_label_position('top') 
ax[0][1].xaxis.set_label_position('top') 
ax[0][2].xaxis.set_label_position('top') 
ax[0][0].set_xlabel('High', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)
ax[0][1].set_xlabel('Middle', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)
ax[0][2].set_xlabel('Low', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)

ax[0][0].set_ylabel('Symmetric', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)
ax[1][0].set_ylabel('Asymmetric', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)

#plt.savefig(os.path.join(FOLDER_FIG, 'objects_row.png'), dpi=300, bbox_inches='tight')
```

```python
# Object Figure - Full-page row version
fig, ax = plt.subplots(1, IMGS.shape[1]*IMGS.shape[0], figsize=(12, 3), layout='compressed')

obj_num = 1
for imgs in range(0, IMGS.flatten().shape[0]):

        img = plt.imread(os.path.join(FOLDER_IMG, IMGS.flatten()[imgs]))
        ax[imgs].imshow(img)
        ax[imgs].grid(False)
        
        #ax[row, col].axvline(img.shape[0]/2, color='k', ls='--', lw=0.6)
        
        #ax[row, col].text(0.03, 0.9, str(obj_num), transform = ax[row, col].transAxes, horizontalalignment='left', color='k', fontsize=LABEL_SIZE)

        obj_num += 1

# Applies to all subplots
for axs in ax.flatten():
    axs.xaxis.set_label_position('top') 
    axs.set_xticklabels([])
    axs.set_xticks([])
    axs.set_yticks([])
    axs.set_yticklabels([])
        
    for axis in ['top','bottom','left','right']:
        axs.spines[axis].set_linewidth(OUTLINE_WIDTH)

ax[0].set_xlabel('A. High / Sym', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)
ax[1].set_xlabel('B. High / Asym', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)
ax[2].set_xlabel('C. Middle / Sym', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)
ax[3].set_xlabel('D. Middle / Asym', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)
ax[4].set_xlabel('E. Low / Sym', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)
ax[5].set_xlabel('F. Low / Asym', fontsize=LABEL_SIZE, labelpad=LABEL_PAD)


plt.savefig(os.path.join(FOLDER_FIG, 'Fig2_Objects_Row.png'), dpi=300, bbox_inches='tight')
plt.savefig(os.path.join(FOLDER_FIG, 'Fig2_Objects_Row.tif'), dpi=300, bbox_inches='tight')
plt.savefig(os.path.join(FOLDER_FIG, 'Fig2_Objects_Row.eps'), dpi=300, bbox_inches='tight')
```

```python
# Task screenshot figure
fig, ax = plt.subplots(1, 4, figsize=(22, 5), dpi=300, layout='compressed')

for col in range(0, TASKS.shape[0]):
    img = plt.imread(os.path.join(FOLDER_IMG, TASKS[col]))
    ax[col].imshow(img)
    textcolor = 'k'
    if col < 2:
        textcolor='w'
    ax[col].text(0.03, 0.9, TASK_LABELS[col], transform = ax[col].transAxes, horizontalalignment='left', color=textcolor, fontsize=19)

for axs in ax:
    axs.grid(False)
    for axis in ['top','bottom','left','right']:
        axs.spines[axis].set_linewidth(OUTLINE_WIDTH)
        axs.spines[axis].set_color('k')
        axs.set_xticks([])
        axs.set_yticks([])
        axs.set_xticklabels([])
        axs.set_yticklabels([])

plt.savefig(os.path.join(FOLDER_FIG, 'Fig1_Methods_Tasks.png'), dpi=300, bbox_inches='tight')
plt.savefig(os.path.join(FOLDER_FIG, 'Fig1_Methods_Tasks.tif'), dpi=300, bbox_inches='tight')
plt.savefig(os.path.join(FOLDER_FIG, 'Fig1_Methods_Tasks.eps'), dpi=300, bbox_inches='tight')
```
